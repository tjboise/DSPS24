```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
file_path = '../DSPS24-1/data/train.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the dataframe to understand its structure
data.head()
data['PCI'].hist(bins=50)
plt.yscale('log')

num_pci_smaller_than_zero = (data['PCI'] == 0).sum()
num_pci_smaller_than_zero

negative_pci_rows = data[data['PCI'] == 0]

# Display these rows
print(negative_pci_rows)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Setting the size of the plot
plt.figure(figsize=(15, 6))

# Creating a box plot
sns.boxplot(x='Section', y='PCI', data=data)

# Rotating the x labels for better visibility
plt.xticks(rotation=45, ha='right')

# Setting titles and labels
plt.title('Relationship between Section and PCI')
plt.xlabel('Section')
plt.ylabel('PCI')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Setting the size of the plot
# plt.figure(figsize=(15, 6))

# Creating a box plot
sns.scatterplot(x='Section', y='PCI', data=data)

# Rotating the x labels for better visibility
# plt.xticks(rotation=45, ha='right')

# Setting titles and labels
plt.title('Relationship between Section and PCI')
plt.xlabel('Section')
plt.ylabel('PCI')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Grouping by 'Section' and calculating mean PCI for each group
plt.figure(figsize=(30, 6))
grouped_data = data.groupby('Section')['PCI'].mean().reset_index()

# Sorting the data for better visualization
grouped_data = grouped_data.sort_values(by='PCI')

# Creating a bar plot to show the distribution of average PCI in each section
plt.figure(figsize=(15, 6))
# sns.barplot(x='Section', y='PCI', data=grouped_data, orient='h')
sns.barplot(y='Section', x='PCI', data=grouped_data, orient='h')

# Rotating the x labels for better visibility
plt.xticks(rotation=45, ha='right')

# Setting titles and labels
plt.title('Average PCI per Section')
plt.ylabel('Section')
plt.xlabel('PCI')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Grouping by 'Section' and calculating mean and standard deviation of PCI for each group
grouped_data_stats = data.groupby('Section')['PCI'].agg(['mean', 'std']).reset_index()

# Sorting the data for better visualization
grouped_data_stats = grouped_data_stats.sort_values(by='mean')

# Creating a vertical bar plot with error bars showing the standard deviation
plt.figure(figsize=(6, 15))
sns.barplot(y='Section', x='mean', data=grouped_data_stats, orient='h', xerr=grouped_data_stats['std'])

# Setting titles and labels
plt.title('Average Damage Index (PCI) per Location (Section) with Standard Deviation')
plt.ylabel('Section')
plt.xlabel('Average Damage Index (PCI)')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

def clean_section_name(section_name):
    return '_'.join(section_name.split('_')[:-1])


train_file_path = '../DSPS24-1/data/train.csv'  # Replace with your train dataset path
train_data = pd.read_csv(train_file_path)

train_data['Section'] = train_data['Section'].apply(clean_section_name)

# Grouping the train data by 'Section' and calculating mean, standard deviation, and count of PCI
train_grouped_data = train_data.groupby('Section')['PCI'].agg(['mean', 'std', 'count']).reset_index()
train_grouped_data.rename(columns={'mean': 'train_PCI_mean', 'std': 'train_PCI_std', 'count': 'train_count'}, inplace=True)

# Load the test dataset
test_file_path = '../DSPS24-1/data/test.csv'  # Replace with your test dataset path
test_data = pd.read_csv(test_file_path)
test_data['Section'] = test_data['Section'].apply(clean_section_name)

# Grouping the test data by 'Section' and calculating mean, standard deviation, and count of PCI
test_grouped_data = test_data.groupby('Section')['PCI'].agg(['mean', 'std', 'count']).reset_index()
test_grouped_data.rename(columns={'mean': 'test_PCI_mean', 'std': 'test_PCI_std', 'count': 'test_count'}, inplace=True)

# Merging the train and test grouped data
combined_data = pd.merge(train_grouped_data, test_grouped_data, on='Section', how='outer')
combined_data.to_csv('../DSPS24-1/data/train-test-compare.csv', index=False)
# Displaying the combined dataframe
print(combined_data.head())  # This displays the first few rows. Remove or adjust for full data.
```
